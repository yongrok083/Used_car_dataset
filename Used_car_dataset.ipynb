```python
import pandas as pd
```

```python
car = pd.read_csv('cars.csv')
```

```python
brand = pd.read_csv('brand.csv')
```

```python
car['brand'] = car['title'].apply(lambda x: x.split()[0].lower())
```

```python
brand['title'] =brand['title'].str.lower()
```

```python
car.head()  # Display the first few rows of the DataFrame
```

```python
brand.rename(columns={'barand': 'brand'}, inplace=True)
```

```python
brand.head()  
```

```python
car_df = car.merge(brand, on='brand')
```

```python
car_df.head(20           )
```

```python
car_df.isna().mean()
```

```python
car_df['Service history'].fillna('Unknown',inplace=True)
```

```python
car_df['Service history'].reset_index()
```

```python
pd.set_option('display.max_columns', None)
```

```python
car_df.describe()
```

```python
car_df = car_df.dropna(subset=['Engine'])
```

```python
car_df
```

```python
car_df.isna().mean()
```

```python
car_df = car_df.dropna(subset=['Emission Class'])
```

```python
car_df.isna().mean()
```

```python
car_df = car_df.dropna(subset=['Seats'])
```

```python
car_df.isna().mean()
```

```python
car_df['Previous Owners'] = car_df['Previous Owners'].fillna(0.0)
```

```python
car_df.isna().mean()
```

```python
car_df =pd.get_dummies(
    car_df,
    columns=['Fuel type', 'Body type', 'Gearbox', 'Emission Class', 'Service history'],
    drop_first=True 
)
```

```python
print(car_df.dtypes)
```

```python
for col in car_df.columns:
    if car_df[col].dtype == 'bool':
        car_df[col] = car_df[col].astype(int)
```

```python
print(car_df.dtypes)
```

```python
bool_cols = car_df.select_dtypes(include='bool').columns
car_df[bool_cols] = car_df[bool_cols].astype(int)
```
